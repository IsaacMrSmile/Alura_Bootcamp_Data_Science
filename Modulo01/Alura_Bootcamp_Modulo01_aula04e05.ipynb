Podemos também pegar uma lista aleatória de cinco estados e plotar o gráfico deles. Podemos fazer isso de várias formas. Uma delas é usar o comando sample() que retorna uma seleção aleatória de n linhas do nosso data frame:

```python
dados.sample(5, random_state= 7)
```

O parâmetro random_state fixa a semente de aleatoriadade, logo o resultado sempre será o mesmo, quando rodar o código. Sem esse parâmetro, a seleção mudaria a cada vez que o código fosse executado. 
Vamos agora plotar o gráfico:

```python
ax = dados.sample(5, random_state= 7).T.plot(figsize = (16,8))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.legend(loc='upper left', shadow=True) 
plt.title('Evolução temporal dos gastos por estado (5 estados aleatórios)', fontdict = {"fontsize":14, "fontweight":"bold"})
plt.show()
```

Também é possível usar um comando da biblioteca Numpy para definir uma semente aleatória.

```python
np.random.seed(7)
dados.sample(5)
```

Pretendemos calcular a correlação entre essas duas séries temporais.

```python
dados_usaveis.shape, df_dengue.shape
```

Observamos então que os dois data frames que pretendemos usar tem tamanhos distintos. Há vários meses que não temos dados regristrados de dengue em alguns estados, provavelmente. Precisamos inserir eses dados como zero usando o comando fillna(). Para isso, utilizaremos as cópias dos data frames que criamos anteriormente:


```python
df_original.head()
```

```python
dados_dengue_original.head()
```

Vamos reconfigurar os índices:

```python
dados_dengue_original.set_index('Unidade da Federação', inplace= True)
dados_dengue_original.head()
```

```python
df_original.set_index('Unidade da Federação', inplace= True)
df_original.head()
```

Vemos que os dois data frame possuem quantidades distintas de colunas. Isso pois os dados de atendimentos começam em Março de 1992 enquanto que os dados de dengue começam em Julho de 2007. Queremos então considerar apenas os meses em comum. 

```python
colunas_usadas = dados_dengue_original.columns.values.tolist()
```

```python
df_original = df_original[colunas_usadas]
df_original.head()
```

Percebemos então que os dados faltantes não estão nulos, mas indicados por '-'. Não podemos usar o fillna(). Teremos que usar o replace() para tratar esses dados. 

```python
df_original.replace('-', 0, inplace= True)
df_original.head()
```

```python
dados_dengue_original.replace('-', 0, inplace= True)
dados_dengue_original.head()
```

Para nossa análise, vamos criar duas séries temporais: uma com a soma do total de gastos em cada mês e outra com a soma do número de casos em cada, a soma tomada sob os quatro estados da região sudeste. 

```python
atendimento_total_sudeste = df_original.drop(['Região', 'Total'], axis = 1).astype(float)[df_original['Região'] == 'Sudeste'].sum().values
atendimento_total_sudeste
```

```python
casos_dengue_sudeste_total = dados_dengue_original.drop(['Região', 'Total'], axis = 1).astype(int)[df_original['Região'] == 'Sudeste'].sum().values
casos_dengue_sudeste_total
```

Agora estamos prontos para calcular a correlação. Para o caso de arrays, podemos fazer isso utilizando o comando corrcoef() do Numpy:

```python
np.corrcoef(casos_dengue_sudeste_total,atendimento_total_sudeste)
```

Vemos então que essas duas séries  temporais tem correlação quase zero. Concluímos que não há relação entre o número de casos de dengue na região Sudeste e os gastos do SUS na mesma região. 

# Aula 04: Análise de gráficos e criação de hipóteses

A maneira como escolhemos plotar um gráfico faz toda a diferença, pois a visualização do gráfico afeta a interpretação que teremos dos dados. Em escalas ou janelas de tempo diferente, podemos tirar interpretações diferente, e provavelmente errôneas. A maneira mais correta e segura seria "ler" e não apenas "ver" o gráfico, mas isso custa um pouco mais de atenção e energia mental. A mossa reação natural é tirar conclusões rápidas a partir da informação visual que recebemos. Por isso uma boa representação visual é importante para passar a mensagem correta. 

```python
ax = dados_usaveis[dados_usaveis['Região'] == 'Sudeste'].drop('Total', axis = 1).T[50:].plot(figsize = (16,8))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.legend(loc='upper center', shadow=True)
#plt.grid(axis = 'y', linestyle = '--', linewidth = 1) 
plt.title('Evolução temporal dos gastos da Região Sudeste', fontdict = {"fontsize":16, "fontweight":"bold"})
plt.xlabel('Tempo (Fer/2008 - Mar/2021)', fontdict= {'fontsize': 13, 'color': 'b'})
plt.ylabel('Gastos', fontdict= {'fontsize': 13, 'color': 'b'})
plt.show()
```

```python
ax = dados_usaveis[dados_usaveis['Região'] == 'Sudeste'].drop('Total', axis = 1).T[50:].plot(figsize = (16,8))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.legend(loc='upper center', shadow=True)
#plt.grid(axis = 'y', linestyle = '--', linewidth = 1) 
plt.title('Evolução temporal dos gastos da Região Sudeste', fontdict = {"fontsize":16, "fontweight":"bold"})
plt.xlabel('Tempo (Fer/2008 - Mar/2021)', fontdict= {'fontsize': 13, 'color': 'b'})
plt.ylabel('Gastos', fontdict= {'fontsize': 13, 'color': 'b'})
plt.ylim([10000000,600000000])
plt.show()
```

## Análise per capita

No entando, comparar os estados entre si pelo valor total gasto em cada mês pode não ser a melhor métrica, já que é esperado que estados mais populosos (e.g., São Paulo) tenham maior gasto. Assim, parece mais razoável fazer essa comparação pelo valor gasto per capita. 

O arquivo csv abaixo com as informações populacionais dos estados entre 2008 e 2020 foi extraído do site do IBGE. Não havia a opção de baixar os dados em .csv, então fiz o download no formato .xls e converti. Mas aproveitei a deixa que os dados estavam em uma planilha e já fiz um tratamento inciais, ainda como planilha: eliminei cabeçalho e rodapé, converti os números para o formato americano e excluí as colunas com o nome das regiões. 
Vou disponibilizar o arquivo tratado no repositório do Github. 

```python
populacao = pd.read_csv('/content/populacao.csv', encoding= 'utf-8', thousands=',', decimal='.')
populacao = populacao.set_index('Unidades da Federação')
populacao
```

Observe, no entanto, que os dados trazem informações populacionais por ano e os data sets que vínhamos trabalhando trazem as informações por ano e mês. Há duas opções:
- Sumarizar as informações do SUS por ano;
- Inserir colunas novas nos dados do IBGE, com as informações mensais.

Acho mais válido a segunda solução. Mas como não temos as informações populacionais por mês (o que faz sentido, já que o IBGE não realiza o censo mensalmente), eu simplesmente vou inserir nos meses do ano x o mesmo valor do ano x.
Para começar, crio uma lista com as siglas dos meses e uma lista com os anos de 2008 à 2020:

```python
meses = ['Jan','Fev', 'Mar','Abr', 'Mai', 'Jun','Jul','Ago','Set','Out','Nov','Dez']
anos = ['2008','2009','2010','2011','2012', '2013','2014','2015','2016','2017','2018','2019','2020']
```

Agora crio o data frame pop, como explicado acima. Aproveito e já vou coletando o nome dos meses que utilizaremos para filtrar no data frame com as informações do SUS:

```python
pop = pd.DataFrame()
coluna_meses = []
for ano in anos:
  for mes in meses:
    nome = ano + '/'+mes
    coluna_meses.append(nome)
    pop[nome] = populacao[ano]
pop.astype(float)      
```

Agora vamos filtrar o nosso data frame do SUS com esses meses.

```python
dados_filtrados = dados_usaveis[coluna_meses]
dados_filtrados
```

Agora podemos criar um data frame com o gasto per capita.

```python
dados_per_capita = dados_filtrados / pop
dados_per_capita['Região'] = dados_usaveis['Região']
dados_per_capita
```

Agora que temos esse data frame, podemos responder a pergunta: quais estados tem o maior gasto médio per capita?

```python
dados_per_capita.drop('Região', axis = 1).T.mean().sort_values(ascending = False)
```

Olhando os gastos per capita, temos uma história bem diferente de quando olhávamos apenas para os valores absolutos. São Paulo figura agora em quinto lugar, ficando atrás de Paraná, Santa Catarina, Rio Grande do Sul e Minas Gerais. Observa-se então que os estados da Região Sul são os que mais investiram (ou receberam recursos) em saúde pública, no período de 2008 a 2020.

Vamos agora comparar os estados com o meu estado, Ceará. 

```python
dados_per_capita_transpostos = dados_per_capita.drop('Região', axis = 1).T
dados_per_capita_transpostos
```

```python
dados_per_capita_CE = pd.DataFrame()
for nome in estados:
  dados_per_capita_CE[nome]= dados_per_capita_transpostos[nome] / dados_per_capita_transpostos['Ceará'] 
```

```python
dados_per_capita_CE = dados_per_capita_CE.T
dados_per_capita_CE['Região'] = dados_usaveis['Região']
```

```python
dados_per_capita_CE
```

```python
dados_per_capita_CE.drop('Região', axis = 1).T.mean().sort_values(ascending = False)
```

Percebemos então que, analisando os gastos médios per capita, a variação entre o Ceará e os outros estados não é tão grande quanto a variação dos gastos médios em valor absoluto. Relembre que, em valores absolutos, São Paulo gastava 6 vezes mais que o Ceará, enquanto que, per capita, Paraná não gasta nem o dobro.
